# Multigridplots

- facet grid
- joint plot
- pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## A second way to plot facet plots --> FacetGrid

```python
# figure level --> relpot, displot, catplot, lmplot
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'violin', hue = 'sex')
```

```python
# facetplot (add col) --> first way
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'violin', hue = 'sex', col = 'day')
```

```python
# facetplot (add col and row) --> first way
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'violin', hue = 'sex', col = 'day', row = 'time')
```

## FacetGrid

```python
g = sns.FacetGrid(data = tips, col = 'day', row = 'time')
g
```

```python
g = sns.FacetGrid(data = tips, col = 'day', row = 'time')
g.map(sns.violinplot, 'sex', 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'day', row = 'time', hue = 'sex')
g.map(sns.boxplot, 'sex', 'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col = 'day', row = 'time', hue = 'sex')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend()
```

## Plotting Pairwise Relationship (PairGrid Vs Pairplot)

- It helps to generate pairwise relationship.
- Pair automatically detect the numerical columns in your dataset, and then plot scatterplot pair by pair.

## Pairplot

```python
iris
```

```python
sns.pairplot(data = iris)
```

```python
sns.pairplot(data = iris, hue = 'species')
```

## PairGrid

```python
g = sns.PairGrid(data = iris)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris, hue = 'species')
g.map(sns.scatterplot)
```

```python
# map_diag, # map_offdiag
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_upper, map_lower
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data = iris, hue = 'species', vars = ['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

## JointGrid Vs Jointplot

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hist', hue = 'sex')
```

```python
g = sns.JointGrid(data = tips, x='total_bill', y = 'tip')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(data = tips, x='total_bill', y = 'tip')
g.plot(sns.scatterplot, sns.boxplot)
```

```python
g = sns.JointGrid(data = tips, x='total_bill', y = 'tip', hue = 'sex')
g.plot(sns.kdeplot, sns.boxplot)
```
