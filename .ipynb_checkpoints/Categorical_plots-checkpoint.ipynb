# Categorical Plots

## Categorical Scatter Plot

- Stripplot --> Axes level
- Swarmplot

## Categorical Distribution Plots

- Boxplot
- Violinplot

## Categorical Estimate Plot -> for central tendency

- Barplot
- Pointplot
- Countplot

### Figure level function -> `catplot`

```python
import seaborn as sns
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

## Stripplot
- one column should be categorical and one is numerical

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill')
```

```python
# jitter
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = False)
```

```python
# using catplot
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip')
```

```python
# jitter
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip', jitter = 0.1)
```

```python
# hue 
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip', hue = 'sex')
```

## Swarmplot
- good for short data

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'swarm')
```

```python
sns.swarmplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

## Boxplot

<p>A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.</p>

image.png

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'day')
```

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

```python
# single bocplot --> numerical col
sns.boxplot(data = tips, y = 'total_bill')
```

## Violinplot
- violinplot = (Boxplot + KDEplot)

```python
# violinplot
sns.violinplot(data = tips, x = 'day', y = 'total_bill', hue = 'day')
```

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', hue = 'day', kind = 'violin')
```

```python
# hue
sns.violinplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

```python
# split
sns.violinplot(data = tips, x = 'day', y = 'total_bill', hue = 'sex', split = True)
```

## barplot

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill', hue = 'sex')
```

```python
# errorbar
sns.barplot(data = tips, x = 'sex', y = 'total_bill', hue = 'sex', errorbar = None)
```

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker')
```

```python
# estimator
sns.barplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator = 'std')
```

```python
sns.barplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator = 'min')
```

## Pointplot

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker')
```

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator = 'min')
```

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator = 'min', errorbar = None)
```

## Countplot

A special case for the bar plot is when you want to show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable

```python
sns.countplot(data = tips, x = 'sex', hue = 'sex')
```

```python
sns.countplot(data = tips, x = 'sex', hue = 'day')
```

```python
# faceting
sns.catplot(data = tips, x = 'sex', y = 'total_bill', col = 'smoker', kind = 'box', row = 'time', hue = 'sex')
```
